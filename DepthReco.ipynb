```python
zstack_file = "image_matrics_lead_bricks_gaps.csv"
```

```python
I_matrices = np.loadtxt(zstack_file)
```

```python
len(I_matrices[0])
```

```python
def generateZStack(I_matrices, interp, returnObj=False):
    imObjs = []
    for i in I_matrices:
        i = i.reshape(21, 21)
        fig = plt.figure(frameon=False)
        ax = plt.Axes(fig, [0., 0., 1., 1.])
        ax.set_axis_off()
        fig.add_axes(ax)
        imshowobj = ax.imshow(np.flip(i), aspect='auto', interpolation=interp)
        imshowobj.set_clim(0.9, 1.2)
        if not returnObj:
            plt.show()
        else:
            imObjs.append(imshowobj)
    if returnObj:
        return imObjs
    
def getImageMatrices(I_matrices):
    im = []
    size = int(len(I_matrices[0])**0.5)
    for i in I_matrices:
        i = i.reshape(size, size)
        im.append(i)
    return im


def getRatioPlot3DforZStack(images):
    for image in images:
        xsize, ysize = image.shape
        x = np.arange(0, xsize, 1)
        y = np.arange(0, ysize, 1)

        xs, ys = np.meshgrid(x, y)
        # z = calculate_R(xs, ys)
        zs = np.array(image)

        fig = plt.figure()
        ax = Axes3D(fig)
        ax.plot_surface(xs, ys, zs, rstride=1, cstride=1, cmap='viridis')
        plt.show()

def getRatioPlot3D(images, index):
        xsize, ysize = images[index].shape
        x = np.arange(0, xsize, 1)
        y = np.arange(0, ysize, 1)

        xs, ys = np.meshgrid(x, y)
        # z = calculate_R(xs, ys)
        zs = np.array(images[index])

        fig = plt.figure()
        ax = Axes3D(fig)
        ax.plot_surface(xs, ys, zs, rstride=1, cstride=1, cmap='viridis',linewidth=0,antialiased=False)
        #ax.plot_wireframe(xs, ys, zs,  rstride=1, cstride=1, cmap='viridis')
        plt.show()
```
